```python
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

# Concatenate todas las columnas normalizadas con las columnas originales 'customerID' y 'Churn'
df_normalizado = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)

display(df_normalizado.head())
```

# Tratamiento de datos

```python
df_normalizado.info()
```

```python
(df_normalizado.isnull().sum())
```

```python
df_normalizado.duplicated()
```

```python
df_normalizado.dtypes
```

```python
categorical_cols = df_normalizado.select_dtypes(include=['object', 'category']).columns
print("\n Inconsistencias en variables categóricas:")
for col in categorical_cols:
    print(f"\n{col}:")
    print(df_normalizado[col].value_counts(dropna=False))
```

```python
print(df_normalizado['Churn'].unique())
```

```python
df_normalizado['Churn'] = df_normalizado['Churn'].replace({'': np.nan, ' ': np.nan, 'Unknown': np.nan})
```

```python
df_normalizado['Churn'] = df_normalizado['Churn'].map({'Yes': 1, 'No': 0})
```

```python
if df_normalizado['Churn'].isnull().sum() > 0:
  df_normalizado['Churn'] = df_normalizado['Churn'].fillna(df_normalizado['Churn'].mode()[0])
```

```python
print(df_normalizado['Churn'].unique())
```

```python
df_normalizado['Charges.Total'] = df_normalizado['Charges.Total'].astype(str).str.strip()
```

```python
df_normalizado['Charges.Total'] = df_normalizado['Charges.Total'].replace({'': np.nan, ' ': np.nan})
```

```python
df_normalizado['Charges.Total'] = pd.to_numeric(df_normalizado['Charges.Total'], errors='coerce')
```

```python
if df_normalizado['Charges.Total'].isnull().sum() > 0:
    print(f"\n {df_normalizado['Charges.Total'].isnull().sum()} valores nulos. Imputando con mediana...")
    median_value = df_normalizado['Charges.Total'].median()
    df_normalizado['Charges.Total'] = df_normalizado['Charges.Total'].fillna(median_value)
```

```python
print(df_normalizado['Charges.Total'].describe())
```

# Cuentas diarias

Esto es un promedio de Cuentas_Diarias = MonthlyCharges / días_del_mes

```python
df_normalizado['Cuentas_Diarias'] = df_normalizado['Charges.Monthly'] / 30
```

```python
print(df_normalizado[['Charges.Monthly', 'Cuentas_Diarias']].head())
```

```python
df_normalizado['Cuentas_Diarias'] = df_normalizado['Cuentas_Diarias'].round(2)
```

```python
df_normalizado.head()
```

```python
df_optimizado = df_normalizado.copy()
```

```python
bool_columns = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
for col in bool_columns:
    df_optimizado[col] = df_optimizado[col].map({'Yes': True, 'No': False})
    df_optimizado[col] = df_optimizado[col].astype('bool')
```

```python
categorical_columns = [
    'gender', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
    'StreamingMovies', 'Contract', 'PaymentMethod'
]
for col in categorical_columns:
    df_optimizado[col] = df_optimizado[col].astype('category')
```

```python
df_optimizado['Churn'] = df_optimizado['Churn'].map({0: False, 1: True}).astype(bool)
```

```python
df_optimizado['SeniorCitizen'] = df_optimizado['SeniorCitizen'].map({0: False, 1: True}).astype(bool)
```

```python
money_columns = ['Charges.Monthly', 'Charges.Total', 'Cuentas_Diarias']
for col in money_columns:
    # Redondear a 2 decimales y convertir a tipo más eficiente
    df_optimizado[col] = df_optimizado[col].round(2).astype('float32')
```

```python
df_optimizado['tenure'] = df_optimizado['tenure'].astype('uint16')
```

# Tratamiento de datos

```python
df_normalizado.info()
```

```python
categorical_cols = df_normalizado.select_dtypes(include=['object', 'category']).columns
print("\n Inconsistencias en variables categóricas:")
for col in categorical_cols:
    print(f"\n{col}:")
    print(df_normalizado[col].value_counts(dropna=False))
```

```python
print(df_normalizado['Churn'].unique())
```

```python
df_normalizado['Churn'] = df_normalizado['Churn'].replace({'': np.nan, ' ': np.nan, 'Unknown': np.nan})
```

```python
df_normalizado['Churn'] = df_normalizado['Churn'].map({'Yes': 1, 'No': 0})
```

```python
if df_normalizado['Churn'].isnull().sum() > 0:
  df_normalizado['Churn'] = df_normalizado['Churn'].fillna(df_normalizado['Churn'].mode()[0])
```

```python
print(df_normalizado['Churn'].unique())
```

```python
df_normalizado['Charges.Total'] = df_normalizado['Charges.Total'].astype(str).str.strip()
```

```python
df_normalizado['Charges.Total'] = df_normalizado['Charges.Total'].replace({'': np.nan, ' ': np.nan})
```

```python
df_normalizado['Charges.Total'] = pd.to_numeric(df_normalizado['Charges.Total'], errors='coerce')
```

```python
if df_normalizado['Charges.Total'].isnull().sum() > 0:
    print(f"\n {df_normalizado['Charges.Total'].isnull().sum()} valores nulos. Imputando con mediana...")
    median_value = df_normalizado['Charges.Total'].median()
    df_normalizado['Charges.Total'] = df_normalizado['Charges.Total'].fillna(median_value)
```

```python
print(df_normalizado['Charges.Total'].describe())
```

## Cuentas diarias

Esto es un promedio de Cuentas_Diarias = MonthlyCharges / días_del_mes

```python
df_normalizado['Cuentas_Diarias'] = df_normalizado['Charges.Monthly'] / 30
```

```python
print(df_normalizado[['Charges.Monthly', 'Cuentas_Diarias']].head())
```

```python
df_normalizado['Cuentas_Diarias'] = df_normalizado['Cuentas_Diarias'].round(2)
```

```python
bool_columns = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
for col in bool_columns:
    df_optimizado[col] = df_optimizado[col].map({'Yes': True, 'No': False})
    df_optimizado[col] = df_optimizado[col].astype('bool')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

palette = sns.color_palette("light:b", n_colors=2)
color_media = palette[0]
color_mediana = palette[1]


variables = ['Meses_Como_Cliente', 'Cargos_Mensuales', 'Cargos_Totales', 'Cuentas_Diarias']
media = [32.35, 64.72, 2279.29, 2.16]
mediana = [29.00, 70.30, 1391.00, 2.34]
desviacion = [24.57, 30.13, 2267.18, 1.00]

x = np.arange(len(variables))
ancho = 0.25

plt.figure(figsize=(14, 8), facecolor='#F5F5F5', )  # Fondo gris claro

# Barras
plt.bar(x - ancho, media, ancho, label='Media', color=color_media, edgecolor='#2c3e50', linewidth=0.1)
plt.bar(x, mediana, ancho, label='Mediana', color=color_mediana, edgecolor='#2c3e50', linewidth=0.1)


for i, (m, d) in enumerate(zip(media, desviacion)):
    #  media
    plt.plot([i - ancho, i - ancho], [m - d/2, m + d/2],
             color='#3498db', linewidth=2.5, alpha=0.8)

    #  mediana
    plt.plot([i, i], [mediana[i] - desviacion[i]/2, mediana[i] + desviacion[i]/2],
             color='#2c3e50', linewidth=2.5, alpha=0.8)

    # Etiquetas con color contrastante
    plt.text(i - ancho, m + d/2 + 50, f'±{d}',
             ha='center', fontsize=10, color='#e74c3c', fontweight='bold')

    plt.text(i, mediana[i] + desviacion[i]/2 + 50, f'±{d}',
             ha='center', fontsize=10, color='#e74c3c', fontweight='bold')


plt.xticks(x, [var.replace('_', ' ') for var in variables], fontsize=12, color='#333333')
plt.ylabel('Valor', fontsize=12, color='#333333')
plt.title('Comparación de Métricas Estadísticas Clave',  fontsize=18, pad=20, color='#333333', fontweight='bold')
plt.legend(fontsize=12, frameon=True, facecolor='white')


plt.grid(axis='y', linestyle='--', alpha=0.3, color='#95a5a6')


plt.yscale('symlog')
plt.yticks(fontsize=10, color='#333333')


plt.axhline(y=0, color='#7f8c8d', linestyle='-', alpha=0.5)


plt.tight_layout()
plt.subplots_adjust(top=0.9)

# Guardar
plt.savefig('metricas_estadisticas_lightb.png', dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```
